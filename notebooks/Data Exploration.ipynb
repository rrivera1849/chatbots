```python
# UDC dataset exploration
dataset_path = '../datasets/twitter_en/'

train = open((os.path.join(dataset_path, 'twitter_en.txt')), 'r')
```

```python
train.seek(0)
max_print = 5

for i, line  in enumerate(train):
    if i % 2 == 0:
        print('>>>> {}'.format(line))
    else:
        print('<<<< {}'.format(line))
    
    if i >= 2 * max_print:
        break
```
